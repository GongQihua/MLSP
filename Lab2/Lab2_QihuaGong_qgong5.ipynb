**Step 1: Write the function *my_poly* and *my_poly_derivative***

The polynomial function is given by

\begin{equation}
f(x)=x^2+5 x-4
\end{equation}

The polynomial and its derivative should be hardcoded in the function *my_poly* and *my_poly_derivative*.


```python
def my_poly(x):
  """
    Argument Descriptions
    Required Input Arguments:
        x: Value in the function domain.

    Required Output Arguments:
        fx: Value of the function at any value x.
    """
  fx = x * x + 5 * x - 4





  #raise NotImplementedError
    
  return fx
```

```python
def my_poly_derivative(x):
  """
    Argument Descriptions
    Required Input Arguments:
        x: Value at where the function can be derivable.

    Required Output Arguments:
        fder: Value of the first derivation of the function at any value x.
    """
  fder = 2 * x + 5





  #raise NotImplementedError
    
  return fder
```

**Step 2: Write the updated function *newtons_update*** based on lecture slides



```python
def newtons_update(x_k):
  """
    Argument Descriptions
    Required Input Arguments:
        x_k: Initial value of x in the current iterative step.

    Required Output Arguments:
        x_kPlus1: Updated value of x in the current iterative step.
    """
  x_kPlus1 = x_k - my_poly(x_k)/my_poly_derivative(x_k)





  #raise NotImplementedError
    
  return x_kPlus1
```

**Step 3: Experiment with different values of xi and Plot**




```python
xi = 10.0

maxIters =  100
tol = 1e-06

# Initialization of relative errors, rel_errs
rel_errs = np.zeros((maxIters,1))

# caluculate function values for each value of xlim_values using for loop
f_values= []
xlim_values = np.arange(-abs(xi), abs(xi), 0.1)
ylim_values = None
for j in xlim_values:
  fvalues = my_poly(j)
  f_values.append(fvalues)
f_values = np.array(f_values)
# plot the xlim_values vs function values and draw x-axis and y-axis centered at origin
# write your code here
#plt.xlim(-abs(xi), abs(xi))
a=plt.gca()
a.spines['right'].set_visible(False)
a.spines['top'].set_visible(False)
a.spines['left'].set_position(('data',0))
a.spines['bottom'].set_position(('data',0))
plt.plot(xlim_values,f_values)
plt.title('xlim_values vs function values')
plt.axis([-15,15,-50,200]) 

# write xi as 'x0' to denote initial point. Use text function to write text on figures
# write from here
plt.text(xi, my_poly(xi), 'x0')

# plot tangent at xi
# write from here
yr_line = my_poly_derivative(xi)*(xlim_values-xi) + my_poly(xi);
plt.plot(xlim_values, yr_line);


# draw line from xi to f(xi).
# write from here
plt.plot([xi,xi],[0, my_poly(xi)])
plt.text(xi,my_poly(xi),'f(10)')




# find Newtons update and write on the same plot
# write from here
x_update = newtons_update(xi)
plt.text(x_update,0,'x1')
plt.show()



```
